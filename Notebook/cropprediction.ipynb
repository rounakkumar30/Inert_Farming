```python
#selecting indpendent feature like nitrogen, phosophorus , postassium, temperature rainfall
#selection dependent features like crop(label)

y = df['label']
x = df.drop(['label'], axis = 1)
```

```python
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score
```

```python
from sklearn.model_selection import train_test_split
#Solitting up the dataset into train and test dataset with ratio of 60:40
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state =0)
```

```python
#Printing the shape of the test and train model

print("The Shape of x train:", x_train.shape)
print("The Shape of x test:", x_test.shape)
print("The Shape of y train:", y_train.shape)
print("The Shape of y test:", y_test.shape)
```

```python
#Using Strafied Cross-Fold Validation for a selection of data in a efficient way

import numpy as np
from sklearn.model_selection import StratifiedKFold

from sklearn.model_selection import cross_val_score


K=StratifiedKFold(n_splits=10)
models = {
          "KNN" : KNeighborsClassifier(),
          "DT" : DecisionTreeClassifier(),
          'RFC' : RandomForestClassifier(),
          'GBC' : GradientBoostingClassifier(),
          'XGB' : XGBClassifier()
          }


model_accuracy = pd.DataFrame(columns=['Model','Accuracy'])


for test, clf in models.items():
    if test == 'XGB' or test=='GBC':
        continue

    result=cross_val_score(clf,x_train,y_train,cv=K)
    print(test)
    print(result)
    print("Minimum value: ",np.min(result))
    print("Maximum value: ",np.max(result))
    print("Average: ",np.mean(result))
    print('\n')

```

```python
import xgboost as xgb
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Convert pandas DataFrames to numpy arrays
x_train_array = x_train.values
y_train_array = y_train.values
x_test_array = x_test.values
y_test_array = y_test.values

# Initialize an empty list to store accuracy values
accuracy_list = []

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform LabelEncoder on the target variable
y_train_encoded = label_encoder.fit_transform(y_train_array)
y_test_encoded = label_encoder.transform(y_test_array)

# Loop through models and calculate accuracy
for test, clf in models.items():
    clf.fit(x_train_array, y_train_encoded)  # Use the encoded target variable
    y_pred = clf.predict(x_test_array)
    acc = accuracy_score(y_test_encoded, y_pred)
    train_pred = clf.predict(x_train_array)
    train_acc = accuracy_score(y_train_encoded, train_pred)
    print("\n", test + ' scores')
    print(acc)
    # Append accuracy values to the list
    accuracy_list.append({'Model': test, 'Accuracy': acc, 'Train_acc': train_acc})

# Convert the list of dictionaries to a DataFrame
model_accuracy = pd.DataFrame(accuracy_list)
```

```python


accuracy_rate=[]
for i in range(1,40):
    
    knn=KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,x_train,y_train,cv=10)
    accuracy_rate.append(score.mean())
```

```python
error_rate=[]
for i in range(1,40):
    
    knn=KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,x_train,y_train,cv=10)
    error_rate.append(1-score.mean())
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('Error rate vs  K Value')
plt.xlabel('K')
plt.ylabel('Error rate')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
RFC=RandomForestClassifier()
params = {
    'n_estimators':[300,400,500],
    'max_depth':[5,10,15],
    'min_samples_split':[2,5,8]
}
grid_rand = GridSearchCV(RFC,params,cv=3,verbose=3,n_jobs=-1)

grid_rand.fit(x_train,y_train)

pred_rand = grid_rand.predict(x_test)

print(classification_report(y_test,pred_rand))

print('Best score : ',grid_rand.best_score_)
print('Best params : ',grid_rand.best_params_)
```

```python
import numpy as np
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform LabelEncoder on the target variable
y_train_encoded = label_encoder.fit_transform(y_train_array)

# Initialize XGBoost classifier
XGB = XGBClassifier()

# Fit the classifier to the training data
XGB.fit(x_train_array, y_train_encoded)

# Encode the test target variable
y_test_encoded = label_encoder.transform(y_test_array)

# Predict on the test data
y_pred = XGB.predict(x_test_array)

# Calculate accuracy
acc = accuracy_score(y_test_encoded, y_pred)
print('Accuracy:', acc * 100)
```
